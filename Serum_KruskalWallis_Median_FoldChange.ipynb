```r
###Loading in feature table and checking length
full.features <- read.csv("Serum_microbiota_elevated_and_depleted.csv")
dim(full.features)
```

```r
###Creating a table of the different donor-diet combinations
table(full.features$ATTRIBUTE_donor_diet.x)
```

```r
###Double-checking where metadata ends and feature abundances begin
head(full.features[,15:16])
```

```r
###Assigning the feature abundances to the new object "noMeta"
noMeta <- full.features[,16:719]
head(noMeta)
```

```r
###Selecting all the features with abundances greater than 0 and assigning
###to the "not_zero" object.
not_zero <- which(colSums(noMeta) !=0)
head(not_zero)
write.csv(not_zero, "Serum_FullFeatures_noMeta_notZero.csv")
```

```r
###Combining all the non-zero metabolite features into the "nonzero" object
nonzero <- noMeta[, not_zero]
head(nonzero)
```

```r
###Attaching metadata
nonzero_wMeta <- cbind(full.features[,1:15], nonzero)
head(nonzero_wMeta)
```

```r
###Checking dataframe lengths to make sure the previous filtering and
###merging steps worked as expected. Also indicates whether columns were lost
dim(full.features)
dim(noMeta)
dim(nonzero)
dim(nonzero_wMeta)
```

```r
###Running a Kruskal-Wallis test based on donor-diet groups and assigning
###those results to the new object "pval".
pval <- c()
for (i in 16:ncol(nonzero_wMeta))
pval <- c(pval, kruskal.test(full.features[,i]~full.features$ATTRIBUTE_donor_diet.x)$p.value) 
```

```r
head(pval)
```

```r
###Performing FDR on our results and assugbubg ti bew ibhect
pAdj <- p.adjust(pval, method=c("fdr"), n=length(pval))
head(pAdj)
```

```r
length(pval)
length(pAdj)
```

```r
pAdj
```

```r
###Saving results as csv
write.csv(pAdj, "Serum_pAdj_for_backup.csv")
```

```r
###Now checking the median of values, assigning to new object, and saving as csv
###This represents medians of fold-changes
mdn <- apply(full.features[,16:ncol(full.features)], 2, function(x) tapply(x, full.features$ATTRIBUTE_donor_diet.x, function(x) median(x, na.rm=TRUE)))
dim(mdn)
write.csv(mdn, "Serum_mdn_for_backup.csv")                                                                           
```

```r
mdn[,1:15]
```

```r
###For each possible donor-diet pair, calculating fold-change based on
###median values. 36 possible pairs
fch_147L_147V <- apply(mdn, 2, function(x) x[1]/x[2])
fch_147L_147WG <- apply(mdn, 2, function(x) x[1]/x[3]) 
fch_147L_165L <- apply(mdn, 2, function(x) x[1]/x[4])
fch_147L_165V <- apply(mdn, 2, function(x) x[1]/x[5])
fch_147L_165WG <- apply(mdn, 2, function(x) x[1]/x[6])
fch_147L_291L <- apply(mdn, 2, function(x) x[1]/x[7])
fch_147L_291V <- apply(mdn, 2, function(x) x[1]/x[8])
fch_147L_291WG <- apply(mdn, 2, function(x) x[1]/x[9])
fch_147V_147WG <- apply(mdn, 2, function(x) x[2]/x[3])
fch_147V_165L <- apply(mdn, 2, function(x) x[2]/x[4])
fch_147V_165V <- apply(mdn, 2, function(x) x[2]/x[5])
fch_147V_165WG <- apply(mdn, 2, function(x) x[2]/x[6])
fch_147V_291L <- apply(mdn, 2, function(x) x[2]/x[7])
fch_147V_291V <- apply(mdn, 2, function(x) x[2]/x[8])
fch_147V_291WG <- apply(mdn, 2, function(x) x[2]/x[9])
fch_147WG_165L <- apply(mdn, 2, function(x) x[3]/x[4])
fch_147WG_165V <- apply(mdn, 2, function(x) x[3]/x[5])
fch_147WG_165WG <- apply(mdn, 2, function(x) x[3]/x[6])
fch_147WG_291L <- apply(mdn, 2, function(x) x[3]/x[7])
fch_147WG_291V <- apply(mdn, 2, function(x) x[3]/x[8])
fch_147WG_291WG <- apply(mdn, 2, function(x) x[3]/x[9])
fch_165L_165V <- apply(mdn, 2, function(x) x[4]/x[5])
fch_165L_165WG <- apply(mdn, 2, function(x) x[4]/x[6])
fch_165L_291L <- apply(mdn, 2, function(x) x[4]/x[7])
fch_165L_291V <- apply(mdn, 2, function(x) x[4]/x[8])
fch_165L_291WG <- apply(mdn, 2, function(x) x[4]/x[9])
fch_165V_165WG <- apply(mdn, 2, function(x) x[5]/x[6])
fch_165V_291L <- apply(mdn, 2, function(x) x[5]/x[7])
fch_165V_291V <- apply(mdn, 2, function(x) x[5]/x[8])
fch_165V_291WG <- apply(mdn, 2, function(x) x[5]/x[9])
fch_165WG_291L <- apply(mdn, 2, function(x) x[6]/x[7])
fch_165WG_291V <- apply(mdn, 2, function(x) x[6]/x[8])
fch_165WG_291WG <- apply(mdn, 2, function(x) x[6]/x[9])
fch_291L_291V <- apply(mdn, 2, function(x) x[7]/x[8])
fch_291L_291WG <- apply(mdn, 2, function(x) x[7]/x[9])
fch_291V_291WG <- apply(mdn, 2, function(x) x[8]/x[9])                        
```

```r
###Combining the fold-change results into a single data frame
df <- data.frame(fch_147L_147V, fch_147L_147WG, fch_147L_165L, fch_147L_165V,
fch_147L_165WG, fch_147L_291L, fch_147L_291V, fch_147L_291WG, fch_147V_147WG,
fch_147V_165L, fch_147V_165V, fch_147V_165WG, fch_147V_291L, fch_147V_291V,
fch_147V_291WG, fch_147WG_165L, fch_147WG_165V, fch_147WG_165WG, fch_147WG_291L,
fch_147WG_291V, fch_147WG_291WG, fch_165L_165V, fch_165L_165WG, fch_165L_291L,
fch_165L_291V, fch_165L_291WG, fch_165V_165WG, fch_165V_291L, fch_165V_291V,
fch_165V_291WG, fch_165WG_291L, fch_165WG_291V, fch_165WG_291WG, fch_291L_291V,
fch_291L_291WG, fch_291V_291WG, pval, pAdj)
head(df)
write.csv(df, "Serum_Df_Merged_Backup.csv")
```

```r
###Subsetting only statistically significant results
df1 <- df[df$pval<0.05,]
dim(df1)
write.csv(df1, "Seurm_Df1_SubsettedFromDf.csv")
```

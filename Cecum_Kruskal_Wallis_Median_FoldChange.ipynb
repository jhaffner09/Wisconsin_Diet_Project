```r
###Loading feature table
full.features <- read.csv("Cecum_microbiota_elevated_and_depleted.csv")
dim(full.features)
```

```r
###Included this to check how the subsetting affected the table. Did not use this object beyond subset tests.
test <- droplevels(subset(full.features, full.features$ATTRIBUTE_donor_diet.x==c("WLS_147_legumes")|full.features$ATTRIBUTE_donor_diet.x==c("WLS_147_vegetable")|full.features$ATTRIBUTE_donor_diet.x==c("WLS_147_whole_grains")|full.features$ATTRIBUTE_donor_diet.x==c("WLS_165_legumes")|full.features$ATTRIBUTE_donor_diet.x==c("WLS_165_vegetable")|full.features$ATTRIBUTE_donor_diet.x==c("WLS_165_whole_grains")|full.features$ATTRIBUTE_donor_diet.x==c("WLS_291_legumes")|full.features$ATTRIBUTE_donor_diet.x==c("WLS_291_vegetable")|full.features$ATTRIBUTE_donor_diet.x==c("WLS_291_whole_grains")))
```

```r
table(full.features$ATTRIBUTE_donor_diet.x)
table(test$ATTRIBUTE_donor_diet.x)
```

```r
###As with lines 2-3, the "test" object was just for comparisons. From here on out, I only use the full "full.features" object.
###Double-checking which columns have feature abundances
head(full.features[,15:16])
head(test[,15:16])
```

```r
###Assigning feature abundances to "noMeta"
noMeta <- full.features[,16:3016]
head(noMeta)
```

```r
###Filtering out any columns where metabolite features have a sum of 0
not_zero <- which(colSums(noMeta) !=0)
head(not_zero)
write.csv(not_zero, "FullFeatures_noMeta_notZero.csv")
```

```r
###Merging the two dataframes into the "nonzero" object
nonzero <- noMeta[, not_zero]
head(nonzero)
```

```r
###Adding metadata back in
nonzero_wMeta <- cbind(full.features[,1:15], nonzero)
head(nonzero_wMeta)
```

```r
###Checking these dimensions helps confirm the previous steps worked
dim(full.features)
dim(noMeta)
dim(nonzero)
dim(nonzero_wMeta)
```

```r
###Creating the object "pval" to contain the results from a Kruskal-Wallis 
###test based on the donor-diet groups.
pval <- c()

for (i in 16:ncol(nonzero_wMeta))
pval <- c(pval, kruskal.test(full.features[,i]~full.features$ATTRIBUTE_donor_diet.x)$p.value)
```

```r
head(pval)
```

```r
###Running FDR correction and adding to a new object "pAdj"
pAdj <- p.adjust(pval, method=c("fdr"), n=length(pval))
head(pAdj)
```

```r
length(pval)
length(pAdj)
```

```r
pAdj
```

```r
###Outputting results to a csv file
write.csv(pAdj, "pAdj_for_backup.csv")
```

```r
###Getting median values from all features, saving as new object, and outputting
###as a csv.
mdn <- apply(full.features[,16:ncol(full.features)], 2, function(x) tapply(x, full.features$ATTRIBUTE_donor_diet.x, function(x) median(x, na.rm=TRUE)))
dim(mdn)
write.csv(mdn, "mdn_for_backup.csv")                                                                            
```
